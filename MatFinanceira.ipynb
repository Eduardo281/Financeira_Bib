```python
import math
from decimal import Decimal

def FVsimples(PV, i, n):
    PV = Decimal(str(PV))
    i = Decimal(str(i))
    n = Decimal(str(n))
    return PV * (1 + i*n)

def PVsimples(FV, i, n):
    FV = Decimal(str(FV))
    i = Decimal(str(i))
    n = Decimal(str(n))
    return FV / (1 + i*n)

def N_simples(FV, PV, i):
    PV = Decimal(str(PV))
    FV = Decimal(str(FV))
    i = Decimal(str(i))
    return ((FV / PV) - 1)/i

def I_simples(FV, PV, n, digitos=5):
    PV = Decimal(str(PV))
    FV = Decimal(str(FV))
    n = Decimal(str(n))
    i = ((FV / PV) - 1)/n
    i = i.quantize(Decimal(str(10 ** -digitos)))
    return i

def FVcomposto(PV, i, n):
    PV = Decimal(str(PV))
    i = Decimal(str(i))
    n = Decimal(str(n))
    return PV * ( (1 + i) ** n)

def PVcomposto(FV, i, n):
    FV = Decimal(str(FV))
    i = Decimal(str(i))
    n = Decimal(str(n))
    return FV * ( 1 / ( (1 + i) ** n) )

def N_composto(FV, PV, i):
    PV = Decimal(str(PV))
    FV = Decimal(str(FV))
    i = Decimal(str(i))
    return Decimal.log10(FV/PV) / Decimal.log10(1 + i)

def I_composto(FV, PV, n, digitos=5):
    PV = Decimal(str(PV))
    FV = Decimal(str(FV))
    n = Decimal(str(n))
    i = (FV/PV) ** (1/n) - 1
    i = i.quantize(Decimal(str(10 ** -digitos)))
    return i

def imprimeFVsimples(PV, i, n):
    PV = Decimal(str(PV))
    i = Decimal(str(i))
    n = round(abs(n))
    parc = PV * i
    aux = PV + parc
    print("-----")
    print("Valores por período: (Opção: Juros Simples)")
    print("---")
    print("Valor Inicial: ${}".format(PV))
    for idx in range(1, n+1):
        print("Período {}: ${}".format(idx, aux))
        aux = aux + parc
    print("-----")
    
def imprimeFVcomposto(PV, i, n):
    PV = Decimal(str(PV))
    i = Decimal(str(i))
    n = round(abs(n))
    aux = PV * (1+i)
    print("-----")
    print("Valores por período: (Opção: Juros Compostos)")
    print("---")
    print("Valor Inicial: ${}".format(PV))
    for idx in range(1, n+1):
        print("Período {}: ${}".format(idx, aux))
        aux = aux * (1+i)
    print("-----")
```

## Matemática Financeira 2:

### Pagamentos Parcelados a Juros Compostos e Montante Acumulado após uma sequência de depósitos parcelados de maneira uniforme

#### Cabeçalhos das Funções:

def PVparcelas(PMT, i, n)

def FVparcelas(PMT, i, n)

def PMT_PV_parcelas(PV, i, n)

def PMT_FV_parcelas(FV, i, n)

```python
import math
from decimal import Decimal

def PVparcelas(PMT, i, n):
    PMT = Decimal(str(PMT))
    i = Decimal(str(i))
    n = Decimal(str(n))
    aux = (1+i) ** n
    return PMT * (  (aux - 1)/(i * aux)  )

def FVparcelas(PMT, i, n):
    PMT = Decimal(str(PMT))
    i = Decimal(str(i))
    n = Decimal(str(n))
    aux = (1+i) ** n
    return PMT * (  ( (1+i) ** n - 1 ) / i  )

def PMT_PV_parcelas(PV, i, n):
    PV = Decimal(str(PV))
    i = Decimal(str(i))
    n = Decimal(str(n))
    aux = (1+i) ** n
    return PV * (  (i * aux) / (aux - 1)  )

def PMT_FV_parcelas(FV, i, n):
    FV = Decimal(str(FV))
    i = Decimal(str(i))
    n = Decimal(str(n))
    return FV * (  i / ( (1+i) ** n - 1)  )

x = float(PVparcelas(1000, 2/100, 8))
x
PMT_PV_parcelas(x, 2/100, 12)

x = 200000
print('Fundos de Renda Fixa:', PMT_FV_parcelas(x, 0.75/100, 120))
i = round( math.pow(1.08, 1/12) - 1, 6 )
print('Poupança:', PMT_FV_parcelas(x, i, 120))
```

## Matemática Financeira 3:

### Amortização de Dívidas: Construção de tabelas de pagamentos pelos Métodos Americano, SAC e Price no formato de DataFrames do Pandas.

#### Cabeçalhos das Funções:

def tabelaAMERICANO(PV, i, n, digitos=QTD_DIGITOS)

def tabelaPRICE(PV, i, n, digitos=QTD_DIGITOS)

def tabelaSAC(PV, i, n, digitos=QTD_DIGITOS)

def gerarTabela(juros, amortizacao, prestacoes, saldo_devedor, digitos=QTD_DIGITOS)


```python
from decimal import Decimal
import numpy as np
import pandas as pd

QTD_DIGITOS = 2

def tabelaAMERICANO(PV, i, n, digitos=QTD_DIGITOS):
    PV = Decimal(str(PV))
    i = Decimal(str(i))
    n = round(abs(n))
    aux = Decimal(str(i * PV))
    juros = [aux]
    prestacoes = [aux]
    juros = juros * (n+1)
    prestacoes = prestacoes * (n+1)
    juros[0] = Decimal(str(0))
    prestacoes[0] = Decimal(str(0))
    prestacoes[n] = PV + aux
    saldo_devedor = [PV] * (n+1)
    saldo_devedor[n] = Decimal(str(0))
    amortizacao = [Decimal(str(0))] * (n+1)
    amortizacao[n] = PV
    tabela = gerarTabela(juros, amortizacao, prestacoes, saldo_devedor, digitos)
    return tabela

def tabelaPRICE(PV, i, n, digitos=QTD_DIGITOS):
    n = round(abs(n))
    prestacoes = [PMT_PV_parcelas(PV, i, n)]
    PV = Decimal(str(PV))
    i = Decimal(str(i))
    prestacoes = prestacoes * (n+1)
    prestacoes[0] = Decimal(str(0))
    saldo_devedor = [PV]
    juros = [Decimal(str(0))]
    amortizacao = [Decimal(str(0))]
    for k in range(1, n+1):
        juros.append(i * saldo_devedor[k-1])
        amortizacao.append(prestacoes[k] - juros[k])
        saldo_devedor.append(saldo_devedor[k-1] - amortizacao[k])
    tabela = gerarTabela(juros, amortizacao, prestacoes, saldo_devedor, digitos)
    return tabela

def tabelaSAC(PV, i, n, digitos=QTD_DIGITOS):
    PV = Decimal(str(PV))
    i = Decimal(str(i))
    n = round(abs(n))
    amortizacao = [PV/n]
    amortizacao = amortizacao * (n+1)
    amortizacao[0] = Decimal(str(0))
    saldo_devedor = [PV]
    juros = [Decimal(str(0))]
    prestacoes = [Decimal(str(0))]
    for k in range(1, n+1):
        juros.append(i * saldo_devedor[k-1])
        prestacoes.append(amortizacao[k] + juros[k])
        saldo_devedor.append(saldo_devedor[k-1] - amortizacao[k])
    tabela = gerarTabela(juros, amortizacao, prestacoes, saldo_devedor, digitos)
    return tabela

def gerarTabela(p_juros, p_amortizacao, p_prestacoes, p_saldo_devedor, digitos=QTD_DIGITOS):
    n = len(p_juros)
    indices = list(range(n))
    indices.append('Total')
    tabela = pd.DataFrame()
    juros = pd.Series(p_juros)
    amortizacao = pd.Series(p_amortizacao)
    prestacoes = pd.Series(p_prestacoes)
    saldo_devedor = pd.Series(p_saldo_devedor)

    juros = juros.apply(lambda x: x.quantize(Decimal(str(10 ** -digitos))))
    amortizacao = amortizacao.apply(lambda x: x.quantize(Decimal(str(10 ** -digitos))))
    prestacoes = prestacoes.apply(lambda x: x.quantize(Decimal(str(10 ** -digitos))))
    saldo_devedor = saldo_devedor.apply(lambda x: x.quantize(Decimal(str(10 ** -digitos))))
 
    juros = juros.append(pd.Series(juros.sum()))
    amortizacao = amortizacao.append(pd.Series(amortizacao.sum()))
    prestacoes = prestacoes.append(pd.Series(prestacoes.sum()))
    saldo_devedor = saldo_devedor.append(
        pd.Series(Decimal(p_saldo_devedor[-1]).quantize(Decimal(str(10 ** -digitos)))))

    tabela['Juros'] = juros
    tabela['Amortização'] = amortizacao
    tabela['Prestação'] = prestacoes
    tabela['Saldo Devedor'] = saldo_devedor
    tabela = tabela.applymap(lambda x: '$ {}'.format(x))
    tabela.index = indices
    
    return tabela

AAA = tabelaAMERICANO(6000, 4.0093/100, 7)
BBB = tabelaPRICE(6000, 4.0093/100, 7)
CCC = tabelaSAC(6000, 4.0093/100, 7)

(tabelaSAC(6000, 4.0093/100, 7))

(tabelaPRICE(50000, 1.5/100, 5))

tabelaPRICE(150000, 0.23, 5)
```

### Análise de Investimento: Valor Presente de uma Sequência de Valores (VP), Valor Presente Líquido (VPL), Taxa Interna de Retorno (TIR) e Tempo de Payback Descontado (PBD).

Estamos realizando o cálculo da Taxa Interna de Retorno através de uma implementação fortemente vetorizada do Método de Newton para o problema. Não vamos entrar em muitos detalhes sobre as fórmulas utilizadas.

```python
import math
import numpy as np
from decimal import Decimal

def sequenciaVP(valores, i):
    # Vamos fazer o cálculo de forma vetorizada por questões de otimização de performance.
    val_np = np.array(valores)
    n = np.size(val_np)
    aux = 1
    x = (1+i)
    pot_x = []
    for idx in range(n):
        pot_x.append(aux)
        aux = aux * x
    pot_x.reverse()
    pot_x = np.array(pot_x)
    aux1 = (val_np * pot_x)/(pot_x[0] * x)
    return aux1

def calculoVPL(PV, valores, i):
    aux = sequenciaVP(valores, i)
    return ( sum(aux) - PV )






def calculoVPL_NOVA(PV, valores, i):
    aux = sequenciaVP_NOVA(valores, i)
    return ( sum(aux) - PV )



def sequenciaVP_NOVA(valores, i):
    # Vamos fazer o cálculo de forma vetorizada por questões de otimização de performance.
    valoresDec = list()
    for item in valores:
        valoresDec.append(Decimal(str(item)))
    i = Decimal(str(i))
    
    val_np = np.array(valoresDec)
    n = np.size(val_np)
    aux = Decimal('1')
    x = (1+i)
    pot_x = []
    for idx in range(n):
        pot_x.append(aux)
        aux = aux * x
    pot_x.reverse()
    pot_x = np.array(pot_x)
    aux1 = (val_np * pot_x)/(pot_x[0] * x)
    return aux1





def calculoTIR(PV, valores, erro = math.pow(10, -10)):
    # Vamos aplicar o Método de Newton tomando x = (1+i)
    val_np = np.array(valores)
    n = np.size(val_np)
    x = 1
    old_x = 1
    epsilon = 1
    while(epsilon >= erro):
        # O primeiro passo é construir o vetor [x^(n-1)*v_1, x^(n-2)*v_2, ..., x*v_{n-1}, v_n].
        # Vamos armazenar este valor na variável aux1
        aux = 1
        pot_x = []
        for idx in range(n):
            pot_x.append(aux)
            aux = aux * x
        pot_x.reverse()
        pot_x = np.array(pot_x)
        aux1 = val_np * pot_x

        # O vetor aux2 é o vetor [x^(n-1)*v_1, 2 * x^(n-2)*v_2, ..., (n-1) * x * v_{n-1}, n * v_n].
        aux2 = np.array(list(range(1, n + 1)))
        aux2 = aux1 * aux2

        # Por fim, a variável aux3 é simplesmente 1 / (x^n):
        aux3 = pot_x[0]*x
        aux3 = 1/aux3

        # O passo do Método de Newton se torna:
        passo_newton = -x * (aux3 * sum(aux1) - PV)/(aux3 * sum(aux2))

        # Atualização do valor de x e cálculo do erro relativo:
        x = x - passo_newton
        epsilon = abs( (x - old_x)/x )
        old_x = x
    # Notando que assumimos que x = (1 + i), retornamos o valor i = x - 1:
    return (x - 1)

def calculoPBD(PV, valores, i, tipo="Anual"):
    valores_presentes = sequenciaVP(valores, i)
    VPL_por_periodo = list()
    aux = 0
    for idx in range(len(valores_presentes)):
        aux = aux + valores_presentes[idx]
        VPL_por_periodo.append(aux - PV)
    if(VPL_por_periodo[-1] < 0):
        print("Valor final negativo!")
        return -1
    for idx in range(len(VPL_por_periodo)):
        if(VPL_por_periodo[idx]) > 0:
            break
    if(idx == 0):
        if(tipo.lower() == "anual"):
            print("Tempo de Payback Descontado: {} dia(s)".format(math.floor( (365 * PV )/(PV + VPL_por_periodo[0]) )))
        if(tipo.lower() == "mensal"):
            print("Tempo de Payback Descontado: {} dia(s)".format(math.floor( (30 * PV )/(PV + VPL_por_periodo[0]) )))
    x = abs(VPL_por_periodo[idx-1])
    if(idx > 0):
        if(tipo.lower() == "anual"):
            print("Tempo de Payback Descontado: {} ano(s) e {} dia(s)".format(idx, math.floor( (365 * x )/(x + VPL_por_periodo[idx]) )))
        if(tipo.lower() == "mensal"):
            print("Tempo de Payback Descontado: {} mês(es) e {} dia(s)".format(idx, math.floor( (30 * x )/(x + VPL_por_periodo[idx]) )))




dados = [254575, 186473, 151353.64, 73979.04, 57185.34]

print("Resultado 1: {}%".format(calculoTIR(PV = 500000, valores = dados) * 100))

dados = [260, 260, 260, 260]
print("Resultado 2: {}%".format(calculoTIR(PV = 1000, valores = dados) * 100))

dados = [2350, 1390, 3350, 4275, 5350]
print("Resultado 3: {}%".format(calculoTIR(PV = 11500, valores = dados) * 100))

print()

dados = [254575, 186473, 151353.64, 73979.04, 57185.34]
print("Resultado VPL 1: ${}".format(calculoVPL(PV = 500000, valores = dados, i=10/100)))

dados = [1000, 1000, 1000, 3000]
print("Resultado VPL 2: ${}".format(calculoVPL(PV = 0, valores = dados, i=8/100)))

dados = [10, 10, 10, 10, 0, 10, 10, 10]
print("Resultado VPL 3: ${}".format(calculoVPL(PV = 0, valores = dados, i=1/100)))

dados = [
    252356, 243212, 241121, 235222, 231596, 220156, 200156, 199548, 198645, 197442, 195843, 193215, 191501,
    189156, 188532, 186487, 183121, 180512, 178465, 177512, 175146, 174561, 173561, 171884, 169004, 168068,
    166135, 164203, 163580, 162513, 161522, 160032, 159486, 158462, 157700, 155046, 154809, 153320, 151809,
    150065, 149807, 148905, 146805, 145555, 143804, 141003, 140068, 139450, 138094, 137705, 135908, 134406,
    133550, 131809, 130055, 129028, 128504, 127806, 126908, 125504, 124804, 123880, 121095, 120650, 119901,
    118900, 117054, 115890, 113577, 112046, 111011, 110974, 110555, 109468, 108555, 107065, 105974, 103804,
    101101, 100024, 
    99530, 98053, 97004, 95855, 94068, 93111, 92028, 90115, 88015, 87755, 85033, 84953, 83133, 82615, 81500,
    80654, 79805, 78808, 77135, 76594, 75488, 74190, 73809, 72115, 70998, 69805, 68708, 67905, 65973, 64063,
    63105, 61809, 60075, 59708, 57806, 56807, 54998, 52898, 51530, 50001
]

print()

print("Tamanho do vetor de dados: {}".format(len(dados)))
print("Resultado TIR com um vetor maior: {}%".format(calculoTIR(PV = 5000000, valores = dados) * 100))

print()

dados = [200, 200, 200, 100, 100, 100]
print("Valores de VP em cada período:")
for j in sequenciaVP(valores=dados, i=1/100):
    print(j)
    
print()
calculoPBD(PV=50000, valores=[15000, 15000, 15000, 15000, 15000], i=8/100)
calculoPBD(PV=80000, valores=[23000, 23000, 23000, 23000, 23000], i=8/100)

print()

calculoPBD(PV=10000, valores=[6000, 5000, 4000, 2500], i=7/100)
calculoPBD(PV=10000, valores=[3000, 4000, 5000, 7000], i=7/100)

print()

calculoPBD(PV=10000, valores=[18000, 30000, 70000, 800000], i=7/100)

print()

calculoPBD(PV=50000, valores=[15000, 15000, 15000, 15000, 15000], i=8/100, tipo="Mensal")
calculoPBD(PV=80000, valores=[23000, 23000, 23000, 23000, 23000], i=8/100, tipo="Mensal")

print()

calculoPBD(PV=10000, valores=[6000, 5000, 4000, 2500], i=7/100, tipo="Mensal")
calculoPBD(PV=10000, valores=[3000, 4000, 5000, 7000], i=7/100, tipo="Mensal")

print()

calculoPBD(PV=10000, valores=[18000, 30000, 70000, 800000], i=7/100, tipo="Anual")

print()

calculoPBD(PV = 10000, valores = [3000, 4000, 5000, 7000], i=10/100)
calculoPBD(PV = 10000, valores = [7000, 5000, 4000, 3000], i=10/100)

dados = [150] * 12
for i in range(12):
    dados.append(200)
for i in range(12):
    dados.append(300)
for i in range(12):
    dados.append(350)
for i in range(12):
    dados.append(400)
for i in range(4):
    dados.append(500)
dados.append(483.69)
    
calculoTIR(PV = 21177.82 - 1894.13, valores = dados)

calculoTIR(PV = 18576.86 - 1894.13, valores = dados)

for i in range(6):
    dados[i] = dados[i] + 1894.13/6
    
calculoTIR(PV = 18576.86, valores = dados)    
```
